```python
import numpy as np
import matplotlib.pyplot as plt
x_arr = np.array([500, 550, 620, 630, 665, 700, 770, 880, 920, 1000]) #size in sq feet
y_arr = np.array([320, 380, 400, 390, 385, 410, 480, 600, 570, 620]) # cost

# Scatter plot, x, y, options. c is color. marker options include:
# c, p, v, . 
plt.scatter(x_arr, y_arr, marker = '.', c = 'b')
# Graph titles
plt.title("Scatter plot of office size and rental price features")
plt.xlabel("Size in Sqaure Feet")
plt.ylabel("Rental price in $")
# plt.show shows the graph. Jupyter does the graph showing for you, so if working in 
# plt.plot does line
plt.show
w = 0.62 # slope
b = 6.47 # y-int

```

```python
# Shape returns type and dimensions, it's a tuple, an immutable list. .shape(0) returns the rows
def compute_model_output (x_train, w, b):
    m = x_train.shape[0]
    y_hat = np.zeros(m) # creates list of size m, all are 0
    for i in range(m):
        y_hat[i] = w * x_train[i] + b # y = mx + b
    return y_hat

y_hat = compute_model_output(x_arr, w, b)
plt.plot(x_arr, y_hat, c = 'r')
plt.scatter(x_arr, y_arr, marker = '.', c = 'b')
plt.title("Scatter plot of office size and rental price features")
plt.xlabel("Size in Sqaure Feet")
plt.ylabel("Rental price in $")
```
